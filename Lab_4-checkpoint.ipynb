```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('AAPL.csv')
dates = pd.DatetimeIndex(df['Date'])

sums = df.groupby(dates.year).sum()

var = 'Volume'
sums[var].plot(kind='pie', fontsize=15)
plt.ylabel(var, horizontalalignment='left')
plt.title('Monthly breakdown for ' + var, fontsize=20)
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets
def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'],
        columns = ds['feature_names'])
    code_species_map = dict(zip(
        range(3), ds['target_names']))
    df['species'] = [code_species_map[c]
        for c in ds['target']]
    return df
df = get_iris_df()

col = 'sepal length (cm)'
df['ind'] = pd.Series(df.index).apply(lambda i: i% 50)
df.pivot('ind','species')[col].plot(kind='box')
plt.title('Petal length (cm)')
plt.show()
```

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('AAPL.csv')

colors = ["r", "g"]
markers= [".", "*"]
fig, ax = plt.subplots(1, 1)
dates = pd.DatetimeIndex(df['Date'])
for i, spec in enumerate(dates.year.unique() ):
    ddf = df[dates.year==spec]
    ddf.plot(kind="scatter",x="Open", y="Close",
             alpha=0.5, s=10*(i+1), ax=ax,color=colors[i], marker=markers[i], 
             label=spec)
    
plt.title('Open vs Close Price')
plt.show()
```
